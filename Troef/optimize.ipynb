```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.layers import Activation, Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Dense, Dropout, Activation, BatchNormalization
from tensorflow.keras.optimizers import SGD, Adagrad, Adam
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.applications.vgg19 import VGG19, preprocess_input, decode_predictions
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from tensorflow.python.keras import activations
from tensorflow.python.ops.nn_ops import dropout
from sklearn.cluster import KMeans

```

```python
data = pd.read_csv("troef.csv",header=None)
print(data.shape)
suits = ["SPADES","DIAMONDS","CLUBS","HEARTS"]

data[8] = data.apply(lambda row: suits.index(row[8]),axis=1)

for i in range(0,32):
    data[f"c{i}"] = data.apply(lambda row: (i in list(row[0:8]))*1,axis=1)

data["y"] = data[8]
data = data.drop([0,1,2,3,4,5,6,7,8],axis=1)
# data = data.drop([8],axis=1)
data.head()
```

```python
train,test = train_test_split(data,test_size=0.2)
x_train = np.array(train.iloc[:, 0:32])
x_test = np.array(test.iloc[:, 0:32])
y_train = np.array(train.iloc[:, -1:])
y_test = np.array(test.iloc[:,-1:])
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```
